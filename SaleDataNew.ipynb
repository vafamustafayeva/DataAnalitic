```python
import pandas as pd
import os
import missingno as msn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path=r"C:\Users\natiq\Desktop\Data Analytics Real-World Projects in Python\Sales_Data"
```

```python
files=os.listdir(path)
print(files)
```

```python
data=pd.DataFrame()
```

```python
for file in files:
    df=pd.read_csv(path+'/'+file)
    data=pd.concat([data, df])
```
