# CLEANING

```python
# "Value" ve "Wage"  sutunlarinda olan deyerlerin deyisdirilmedi
# mes: €110.5M > 110.500000
```

```python
def convert(row):
    row=row.replace('€', '')
    if 'M' in row:
        row=row.replace('M', '')
        row=float(row)*1000000
    elif 'K' in row:
        row=row.replace('K', '')
        row=float(row)*1000
        
    return float(row)
```

```python
data['Wage']=data['Wage'].map(convert)
```

```python
data['Value']=data['Value'].map(convert)
```

```python
data.head()
```

```python
# "Weight" sutununda olan lbs silmek lazimdir s
```

```python
data.loc[data['Weight'].notnull(), 'Weight'] = data.loc[data['Weight'].notnull(),'Weight'].map(lambda row: float(row.replace('lbs',' ')))
```

```python
data.head()
```

```python
# "Height" uzunlugunu ve "Release Clause" duzgun formaya getirilmesi
```

```python
def convert_height (row):
    res= row.split("'")
    return float((float(res[0]))*12+float(res[1])*2.54)
```

```python
data.loc[data['Height'].notnull(), 'Height'] = data.loc[data['Height'].notnull(), 'Height'].map(convert_height)
```

```python
data.loc[data['Release Clause'].notnull(), 'Release Clause'] = data.loc[data['Release Clause'].notnull(), 'Release Clause'].map(convert)
```

```python
data.head()
```

```python
data['Height']=data['Height'].astype(float)
data['Weight']=data['Weight'].astype(float)
data['Release Clause']=data['Release Clause'].astype(float)
```

```python
# Numeric ve Catagoric datalarin gotururlmesi 
```

```python
numeric_cols = data._get_numeric_data().columns
print('Numeric sutun sayi: ', len(numeric_cols))
```

```python
catagoric_cols = list(set(data.columns) -set(numeric_cols))
print('Catagoric sutun sayi: ', len(catagoric_cols))
```

```python
# Nullari dodurmaq 
```

```python
data['Club'].isnull().sum()
```

```python
data['Club'].fillna('No Club', inplace =True)
```

```python
# Numeric datalari Median-la, Catagoric datalari ise Mode ile doldurulmasi
```

```python
for col in numeric_cols:
    data[col].fillna(data[col].median(),inplace=True)
```

```python
for col in catagoric_cols:
    data[col].fillna(data[col].mode()[0],inplace=True) # pandasin mode() serias qaytarir buna gore 0-ci gotururlur
```
